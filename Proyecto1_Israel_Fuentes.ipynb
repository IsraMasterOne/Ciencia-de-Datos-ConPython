# Ejercicio 1


###Escribir un programa que pregunte al usuario por las ventas de los últimos 5 días, y muestre por pantalla una serie con los datos de las ventas por producto: Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%.

```python
#Creamos un diccionario con la lista de articulos y el precio unitario
articulos={"jeans":100,"blusa":70,"playera":50,"sueter":200,"gorra":90}
#Creamos una lista vacía donde añadiremos mas tarde el numero de articulos vendidos,venta total y venta total con descuento
ventas=[]
#Iteramos dos "for" para ir preguntando día por día el numero de ventas de dicho artículo
for i in articulos.keys():
  for j in range(5):
    venta=-1
    while(venta<0):
      print("Introduzca la venta para el día "+str(j+1)+": para el articulo: "+str(i))
      venta=int(input())
      output.clear()
    ventas.append(venta)
    ventas.append(venta*articulos[i])
    ventas.append(venta*articulos[i]*0.95)
```

```python
#Con ayuda de numpy, cambiamos la dimensión de la serie de 1x75 a 5x15
ventas=np.reshape(ventas, (5, 15))
ventas
```

```python
#En una lista creamos las columnas que después separaremos
columnas=["unidades vendidas_Jeans","SinDescuento_Jeans","Descuento_Jeans","unidades vendidas_Blusa","SinDescuento_Blusa","Descuento_Blusa","unidades vendidas_Playera","SinDescuento_Playera","Descuento_Playera","unidades vendidas_Sueter","SinDescuento_Sueter","Descuento_Sueter","unidades vendidas_Gorra","SinDescuento_Gorra","Descuento_Gorra"]
#Creación del Dataframe
df=pd.DataFrame(ventas,columns=columnas)
#Separamos las columnas en subcolumnas
df.columns = df.columns.str.split('_', expand=True)
df.columns = df.columns.swaplevel(1,0)
#Creación del index
df['Dia']= range(1,6)
#Asignamos index
df.set_index('Dia',inplace=True)
df
```

### Escribir un programa que reciba un diccionario con los ingresos mensuales de los clientes de una empresa, y devuelva una serie con el ingreso mínimo, máximo y media.

```python
#Función creada que tiene como especialidad
#recibir un diccionario con la lista de empleados
#y su número de ventas mensuales, devolviendo la
#cifra mas alta, baja, su respectivo vendedor y 
#promedio de ventas
def Statistic_Employee(Empleados):
  for i in range(len(Empleados.keys())):
    ventas_m=int(input("Ingrese la venta mensual del empleado: "+str(list(Empleados.keys())[i])+"--> "))
    #Se modifica en el diccionario el valor de cada llave
    Empleados[str(list(Empleados.keys())[i])]=ventas_m
  #A partir del diccionario se crea un DataFrame
  dfE=pd.DataFrame(Empleados.items(),columns=['Empleado','Ventas mensuales'])
  #Se asigna el index
  dfE.set_index(['Empleado'],inplace=True)
  output.clear()
  #La fucnción devuelve la mayor venta, el index de la mayor venta, la menor, el index de la menor venta y el promedio
  return pd.Series([dfE['Ventas mensuales'].idxmax(),int(dfE.max()),dfE['Ventas mensuales'].idxmin(),int(dfE.min()),float(dfE.mean())])
```

```python
#Prueba de la funcion dos diccionarios distintos
Empleados={'Luis Mario':0,'Ivan':0,'Kimberly':0,'Bruce':0,'Nacho':0}
Gatos={'Lala':0,'Nacho':0,'Cerezo':0}
Eval=Statistic_Employee(Gatos)
print("Empleado de mayor rendimiento: "+Eval[0]+" con ventas: "+str(Eval[1]))
print("Empleado de menor rendimiento: "+Eval[2]+" con ventas: "+str(Eval[3]))
print("Promedio de venta: "+str(Eval[4]))
```

```python
#En una lista se crea los datos a ingresar al dataframe
Datos=[['Enero',30500,22000],['Febrero',35600,23450],['Marzo',28300,18100],['Abril',33900,35700],['Mayo',42500,32450]]
```

```python
#Se crea el DataFrame
df4=pd.DataFrame(Datos,columns=['Mes','Ventas','Gastos'])
#Se añade la columna de Gastos sobre Ventas
df4['Gastos/Ventas*100']=(df4['Gastos']/df4['Ventas'])*100
#Se añade la columna donde si la Venta es mayor a 30000 se coloca Meta Superada, de lo contrario Meta no Superada
df4['Meta']=np.where(df4['Ventas']>= 30000, 'Meta Superada', 'Meta no superada')
#Se añade la columna donde se calcula el sobregasto, de lo contrario no se calcula
df4['Sobre_Gasto']=np.where(df4['Gastos']> df4['Ventas'], abs(df4['Ventas']-df4['Gastos']),"--")
```
